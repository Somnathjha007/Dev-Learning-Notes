```python
#Enclosing Function Loals
name = 'This is a global String'

def greet():
    #Enclosing
    name='I am a local'
    
    def hello():
        #Local
        print('Hello ' + name)
    
    hello()

greet()
        
```

```python
x = 50
def fun():
    global x #Go to global space and connect to it , ass all changes made will effect it
    
    print(f'X is {x}')
    
    #Local Reassignment !
    x = 200
    print(f'I just Locally Changed X to {x}')
```

```python
fun()
print(x)
```
