```python
#map(func, *iterables) --> map objects
#
def square(num):
    return num**2
```

```python
my_nums = [1,2,3,4,5]
```

```python
for item in map(square,my_nums):
    print(item)
```

```python
list(map(square,my_nums))
```

```python
def splicer(myString):
    if len(myString)%2==0:
        return 'EVEN'
    return myString[0]
```

```python
names = ['Andy','eve','sally']
```

```python
list(map(splicer,names))
```

```python
def check_even(num):
    return num%2==0
```

```python
myNums = [1,2,3,4,5,6,7]
```

```python
#
filter(check_even,myNums)
```

```python
list(filter(check_even,myNums))
```

```python
#lambda function is a anon function , which are typically written only to use once
square = lambda num: num **2
```

```python
square(3)
```

```python
list(map(lambda num:num**2,myNums))
```

```python
list(filter(lambda num:num%2==0,myNums))
```
