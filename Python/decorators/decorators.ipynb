```python
# Generators allows us to generate a sequence of value overtime.
```

```python
def create_cubes(n):
    result = []       #using list ot store all of these in memory !!!
    for x in range(n):
        result.append(x**3)
    return result
```

```python
create_cubes(10)
#keeping an entire list in memory.
```

```python
for x in create_cubes(10):
    print (x)
```

```python
def create_cubes2(n):
    for x in range(n):
        yield x**3       #yield is a keyword
   
```

```python
for x in create_cubes2(10):   # same thing , but now more memory efficient 
    print (x)
```
